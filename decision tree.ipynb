```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tensorflow.keras import layers,models
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
#data loading
dataset_paths = []
for i in range(15):
    dataset_paths.append("./dataset/data"+str(i+1)+".csv")

all_datasets = []

for dataset_path in dataset_paths:
    current_dataset = pd.read_csv(dataset_path)
    all_datasets.append(current_dataset)

df = pd.concat(all_datasets, ignore_index=True)
```

```python
df["marker"]=df["marker"].replace({"Natural":0,"Attack":1})
```

```python
df=df.replace([np.inf,-np.inf],np.nan)
```

```python
p = df.isna().sum()
print(p.values)
```

```python
df=df.dropna()
```

```python
df.describe()
```

```python
#segregating independent and dependent varibales using index location menthod
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
 x_train, x_test, y_train, y_test = train_test_split(  x, y, test_size=0.30, random_state=42)
```

```python
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
# Initialize the Decision Tree classifier
clf = DecisionTreeClassifier(max_depth=6, random_state=100)

# Fit the classifier on the training data
clf.fit(x_train, y_train)
```

```python

y_pred = clf.predict(x_train)

# Step 3: Calculate Accuracy
training_accuracy = accuracy_score(y_train, y_pred)
print("Training Accuracy:", training_accuracy)
```

```python
y_predict=clf.predict(x_test)
```

```python
y_pred=[]
for val in y_predict:
    if(val==0):
        y_pred.append(0)
    else:
        y_pred.append(1)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```
